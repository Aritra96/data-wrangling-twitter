#### Identify the dog gender from text:

#### Define

We do not have a column that stores the dog genders. Gender would be an interesting variable to consider for analyses and therefore in this step, we create a new column called 'gender' by programmatically extracting it from the 'text' column.

#### Code

```python
# Creating lists of possible gender pronouns

MALE_PRONOUNS = set(["him", "he's", "his", "he", "himself"])

FEMALE_PRONOUNS = set(["her", "she's", "she", "herself"])
```

```python
# Creating a function to select gender from sentences from 'text' column

def gender_check(words):
    """ This function takes in words from the text column of 
    twitter_archive and determines whether the gender of a dog
    is male or female or both or none.
    
    Args:
    str (words) - takes text/string as input
    
    Returns:
    gender (str) - returns the identified gender as output as string
    
    """
    male_length = len(MALE_PRONOUNS.intersection(words))
    female_length = len(FEMALE_PRONOUNS.intersection(words))

    if male_length > 0 and female_length == 0:
        gender = 'Male'
        
    elif male_length == 0 and female_length > 0: 
        gender = 'Female'
        
    elif male_length > 0 and female_length > 0: 
        gender = 'Both'
        
    else:
        gender = 'None'
        
    return gender
```

```python
# Cleaning the 'text' column prior to applying this function

twitter_archive['text'] = twitter_archive['text'].str.replace('.', '')
twitter_archive['text'] = twitter_archive['text'].str.replace(',', '')
```

```python
# Applying the gender_check function on a newly created column - 'gender'

twitter_archive['gender'] = twitter_archive['text'].str.lower().str.split(" ").apply(gender_check)
```

Some columns have odd dog 'names' such as 'a', 'an', 'the'. In this step, we handle those names. Additionally, some names are in lowercase, so we convert them to titlecase.

```python
# First, create a dictionary to map the articles to more relevant NaN value

name_map = {'a': np.nan, 'an': np.nan, 'the': np.nan, 'None': np.nan}

# Find and replace all articles with NaN

twitter_archive['name'].replace(name_map, inplace=True)

# Capitalize the names that are identified correctly if they are in lowercase

twitter_archive['name'] = twitter_archive['name'].str.title()
```

1. It can be said that retweet_count and favorite_count are linearly related and they seem to follow a positive linear trend. One could say that they are positively correlated.
